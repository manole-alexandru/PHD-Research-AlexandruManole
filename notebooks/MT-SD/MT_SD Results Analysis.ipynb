```python
import pandas as pd
import matplotlib.pyplot as plt

# List of CSV file paths
files = ["/content/drive/MyDrive/prototypes/mtsd_exp/mnist_mt_run1/metrics/val_metrics.csv",
         "/content/drive/MyDrive/prototypes/mtsd_exp/mnist_st_run1/metrics/val_metrics.csv"]

plt.figure(figsize=(10, 6))

for file_id, file in enumerate(files):
    df = pd.read_csv(file)
    if file_id == 0:
        key = "val_mse_x0"
    else:
      key = "val_mse"
    plt.plot(df["epoch"], df[key], label=file)

plt.xlabel("epoch")
plt.ylabel("val_mse_eps")
plt.title("MSE Evolution")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of CSV file paths
files = ["/content/drive/MyDrive/prototypes/mtsd_exp/mnist_mt_run1/metrics/val_metrics.csv",
         "/content/drive/MyDrive/prototypes/mtsd_exp/mnist_st_run1/metrics/val_metrics.csv"]

plt.figure(figsize=(10, 6))

for file in files:
    df = pd.read_csv(file)
    plt.plot(df["epoch"], df["fid"], label=file)

plt.xlabel("epoch")
plt.ylabel("fid")
plt.title("fid Evolution")
plt.legend()
plt.grid(True)
plt.show()
```
